```python
#correlation between Total_Bilirubin and Direct_Bilirubin is high so we can drop any one
#checking variance to choose which one to drop using violin plot
dt = df[["Total_Bilirubin", "Direct_Bilirubin"]]
sns.violinplot(data=dt)
```

```python
#dropping column with low variance because it has less predictive power
df=df.drop(["Direct_Bilirubin"],axis=1)
```

```python
#Converting categorical data("gender") to numeric
df=pd.get_dummies(df,drop_first=True)
```

```python
#Checking dataset after converting gender to numeric column
df.head()
```

```python
#splitting input and output
x=df.drop(["Dataset"],axis=1).values
y=df["Dataset"].values
```

```python
#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.3, random_state = 1234)
```

# scaling the data

```python
#Scaling x with Standard Scaler
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x_train[:,:-1]=s.fit_transform(x_train[:,:-1])
x_test[:,:-1]=s.transform(x_test[:,:-1])
```

```python
#Declaring model name and accuracy,recall rate, f1 score that is needed to be found for later evaluation of model
model=["Logistic regression","Random Forest Classifier","Naive Bayes","K-Neighbor Classifier","Decision Tree Classifier"]
acc=[]
rec=[]
f1=[]
```

# Logistic Regression 

```python
#Training the LogisticRegression model on the Training set
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,accuracy_score,recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
l=LogisticRegression()
l.fit(x_train,y_train)

#predicting output
y_pred=l.predict(x_test)

#Evaluating the model
print("score-train data:", l.score(x_train,y_train))
print("score-test data:", l.score(x_test,y_test))
print("classification report: \n",classification_report(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt="d")
acc.append(accuracy_score(y_test,y_pred))
print("accuracy: \n",accuracy_score(y_test,y_pred))
rec.append(recall_score(y_test, y_pred))
f1.append(f1_score(y_test, y_pred))
```

# Random Forest Classifier

```python
#Training the RandomForestClassifier model on the Training set
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=11,random_state=3)
rf.fit(x_train,y_train)

#predicting output
y_pred=rf.predict(x_test)

#Evaluating the model
print("score-train data:", rf.score(x_train,y_train))
print("score-test data:", rf.score(x_test,y_test))
print("classification report: \n",classification_report(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt="d")
acc.append(accuracy_score(y_test,y_pred))
print("accuracy: \n",accuracy_score(y_test,y_pred))
rec.append(recall_score(y_test, y_pred))
f1.append(f1_score(y_test, y_pred))
```

# Naive Bayes

```python
#Training the Naive Bayes model on the Training set
from sklearn.naive_bayes import GaussianNB
g=GaussianNB()
g.fit(x_train,y_train)

#predicting output
y_pred=g.predict(x_test)

#Evaluating the model
print("score-train data:", g.score(x_train,y_train))
print("score-test data:", g.score(x_test,y_test))
print("classification report: \n",classification_report(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
acc.append(accuracy_score(y_test,y_pred))
print("accuracy: \n",accuracy_score(y_test,y_pred))
rec.append(recall_score(y_test, y_pred))
f1.append(f1_score(y_test, y_pred))
```

# K-Neighbor Classifier

```python
#Training the KNeighborsClassifier model on the Training set
from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier(n_neighbors=7,metric='minkowski', p=2)
k.fit(x_train,y_train)

#predicting output
y_pred=k.predict(x_test)

#Evaluating the model
print("score-train data:", k.score(x_train,y_train))
print("score-test data:", k.score(x_test,y_test))
print("classification report: \n",classification_report(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt="d")
acc.append(accuracy_score(y_test,y_pred))
print("accuracy: \n",accuracy_score(y_test,y_pred))
rec.append(recall_score(y_test, y_pred))
f1.append(f1_score(y_test, y_pred))
```

# Decision Tree Classifier

```python
#Training the DecisionTreeClassifier model on the Training set
from sklearn.tree import DecisionTreeClassifier
d= DecisionTreeClassifier(criterion="entropy",random_state=3)
d.fit(x_train,y_train)

#predicting output
y_pred=d.predict(x_test)

#Evaluating the model
print("score-train data:", d.score(x_train,y_train))
print("score-test data:", d.score(x_test,y_test))
print("classification report: \n",classification_report(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt="d")
acc.append(accuracy_score(y_test,y_pred))
print("accuracy: \n",accuracy_score(y_test,y_pred))
rec.append(recall_score(y_test, y_pred))
f1.append(f1_score(y_test, y_pred))
```

# Comparing performance of models

```python
#Comparing accuracy score of different models
for i in range(0,len(acc)):
    print(model[i],":",round(acc[i],2))
```

```python
#plotting accuracy,Recall rate and f1 score for each model
N = 5
ind = np.arange(5) 
width = 0.25
plt.figure(figsize=(12,5))
#plot for accuracy
bar1 = plt.bar(ind, acc, width, color = 'r')
  
#plot for Recall
bar2 = plt.bar(ind+width, rec, width, color='y')
  
#plot for f1 Score
bar3 = plt.bar(ind+width*2, f1, width, color = 'b')
plt.xticks(ind+width,model)
plt.legend( (bar1, bar2, bar3), ('accuracy', 'recall', 'f1 score') )
plt.show()
```



# Conclusion

As per the above result, although the Random forest classifier shows better result in terms of accuracy, its recall and f1 score are comparatively low than the logistic regression. Since recall rate is very important in this cases because the tolerance of false positive is low, Logistic regression would be better model in this scenario.
